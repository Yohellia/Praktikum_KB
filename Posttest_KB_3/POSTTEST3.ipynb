# **VISUALISASI DATA**

#### Visualisasi Data menggunakan Pandas, Matplotlib, dan Seaborn jadi kita harus import Pandas, Matplotlib, dan Seaborn untuk menggunakannya.

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
```

#### Buka dataset menggunakan command **df=pd.read_csv('Nama_Dataset').**

```python
df = pd.read_csv('PS_20174392719_1491204439457_log.csv')
```

#### Karena data saya banyak kurang lebih 6jt record sehingga saya membuat variable baru yang di dalamnya terdapat sample. Didalam sample ini terdapat data saya, namun saya hanya menggunakan sebanyak n, n yaitu 2000 record.

```python
shelom = df.sample(n=2000)
```

```python
shelom.shape
```

#### atur *step* menjadi index agar mudah menggunakan loc

```python
shelom.set_index('step', inplace=True)
shelom.head()
```

##### 

## **BAR PLOT**

```python
#pandas
transaksi = shelom['type'].value_counts()
transaksi.plot.bar()
plt.title('Perbandingan Jumlah Penipuan Berdasarkan Jenis Transaksi')
```

##### Perbandingan jumlah penipuan berdasarkan jenis transaksi

#### Membuat variable *transaksi* yang di dalamnya penggabungan kolom berdasarkan *type*. Agar memudahkan pemanggilan dalam pemvisualisasian.

```python
transaksi = shelom.groupby('type').sum()
transaksi.head()
```

```python
#matplotlib
plt.bar(transaksi.index,
        transaksi['isFraud'])
plt.xticks(rotation=20, ha='right')
```

##### Dari data berdasarkan table, penipuan banyak terjadi ketika melakukan transaksi *cash_out* dan *transfer*

```python
#seaborn
sns.barplot(x=transaksi.index,
            y=transaksi['isFraud'])
plt.xticks(rotation=60, ha='right')
```

##### Dari data berdasarkan table, penipuan banyak terjadi ketika melakukan transaksi *cash_out* dan *transfer*

## **PIE CHART**

```python
transaksi.index
```

```python
# pandas
explode = [0, 0, 0, 0, 0]
transaksi['isFraud'].sort_values().plot.pie(autopct='%.1f%%', startangle=90, explode=explode)
```

##### penggunaan pie chart untuk mengetahui dalam bentuk persen. Dari data di atas dapat di taraik kesimpulan bahwa terjadinya penipuan karena transaksi *cash_out* dan *transfer*

```python
# matplotlib
plt.pie(transaksi['isFraud'].sort_values(),
        labels=transaksi['isFraud'].sort_values().index,
        autopct='%.1f%%',
        startangle=90,
        explode=explode)
```

##### penggunaan pie chart untuk mengetahui dalam bentuk persen. Dari data di atas dapat di taraik kesimpulan bahwa terjadinya penipuan karena transaksi *cash_out* dan *transfer*

## **HEATMAP**

```python
total_corr = shelom.loc[:,'oldbalanceDest':'isFraud'].corr()
total_corr
```

##### Membuat variable *total_corr* untuk membuat batasan dalam *heatmap*

```python
# pandas
total_corr.style.background_gradient(cmap='coolwarm').set_properties(**{'font-size':'30px'})
```

```python
# matplotlib
plt.imshow(total_corr,cmap='coolwarm')
plt.colorbar()
plt.xticks(range(len(total_corr)), total_corr.columns)
plt.yticks(range(len(total_corr)), total_corr.columns)
plt.grid(False)
```

```python
# seaborn
sns.heatmap(
    data=total_corr,
    annot=True
)
```

##### Korelasi antar *newbalanceDest* dengan *oldbalanceDest* lebih tinggi dari pada *isFraud*.
